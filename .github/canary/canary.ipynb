# QuBins CI canary

Exercises the same surface a tutorial reader hits in their first cell: build a 2-qubit circuit, run it on Aer, assert the histogram has the two expected bitstrings. Run inside every freshly-built image via `jupyter nbconvert --execute` to catch regressions the Dockerfile's `import qiskit` smoke test can't (notebook server, kernel startup, Aer in-process simulation, plot dependencies — pylatexenc etc. on xl). Failing this should block publish.

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1)
qc.measure_all()

sim = AerSimulator()
tqc = transpile(qc, sim)
result = sim.run(tqc, shots=1024).result()
counts = result.get_counts()

assert set(counts.keys()).issubset({'00', '11'}), counts
assert sum(counts.values()) == 1024, counts
print('canary ok:', counts)
```
